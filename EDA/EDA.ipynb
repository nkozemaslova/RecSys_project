**Описание данных с describe**

1. **Disc_number** (номер диска в альбоме) имеет значение в квантилях 1, в среднем по данным композиции находятся в первом диске у исполнителя. Имеется максимальное значение 13, что является выбросом. 
2. Признаки **Dancebility** (насколько трек подходит для танцев); **Energy** (насколько трек интенсивный и активный); **Acousticness** (показатель того, является ли трек акустическим) и **Valence** (насколько негативно (0) или позитивно (1) звучит трек)  -  хорошо распределены от 0 до 1. 
3. **Loudness** (громкость трека в дБ) имеет сильный разброс: минимальное значение -60, максимальное 7.234.
4. **Speechiness** (доля произнесенных слов в треке) 75% квантиль 0.072, при среднем 0.084 -> в данных имеются выбросы
6. **Liveness** (вероятность того, что трек был исполнен вживую) среднее составляет 0,202 - большая часть треков реализованы в записи, а не вживую.
7. **Duration_ms** (длительность трека в миллисекундах) максимальное значение 101 минута (выброс) и минимальное 1 секунда (выброс), стандартное отклонение имеет завышенное значение - 2,7 минуты, значит в данных большой разброс
8. **Year** (год выпуска) минимальное значение 0, что является некорректным представлением данных для этого признака.

**Описание данных с describe**

1. Колонка **Unmaded: 0** дублирует индексы объектов, ее необходимо удалить
2. Признак **Personalized** (тип рекомендации пользователю, где 2 – наиболее подходящая композиция, 1 - подходящая, 0 – случайная) имеет среднее значение 1, то есть процент рекомендаций, предлагаемых системой Spotify редко классифицируется как наиболее подходящий.
3. Признак **Song_id** показывает номер песни в системе Spotify, стандартное отклонение имеет высокое значение -> песни в датасете собраны случайным образом, последовательной выборки нет. 
4. Признак **Spotify_popularity** имеет равномерное распределение, значения представлены от 0 до 100. 
5. Признак **Liked** показывает оценку пользователем композиции, 0 - не понравилась, 1 - нейтрально, 2 - понравилась

**Соотношение наиболее популярных песен (с рейтингом 100) и оценок этих песен пользователями**

Вывод: большинство композиций, имеющих популярность 100 имеют оценку пользователями 1 (нейтрально).

**Соотношение песен с рейтингом меньше 50 и оценок пользователей**

```python
sns.jointplot(data = unpopular_songs, x = 'spotify_popularity', y = 'liked', kind = 'scatter');
```

Вывод: Композиции с рейтингом популярности 50 и меньше чаще всего имеют негативную оценку пользователями и очень редко положительную.
